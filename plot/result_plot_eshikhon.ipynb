```python
labels = ['Area 2', 'Area 4', 'Area 8']
avg_reward_all = [avg_2, 
                  avg_4, 
                  avg_6 
                  ]
x = np.arange(3)
fig, ax = plt.subplots(figsize=(15,13))
width = 0.3 # the width of the bars

rects1 = ax.bar(x + width/2, avg_reward_all, width, label=labels,color=['green', 'blue', 'red'])

ax.set_ylabel('Average AP Selection accuracy [%]',fontsize='18',weight='bold')

plt.legend(rects1[:4], ['Area 2', 'Area 4','Area 8'],loc='upper left');

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
plt.savefig("result/eshikhon2.pdf",dpi=300, bbox_inches='tight')
plt.savefig("result/eshikhon2.png",dpi=300, bbox_inches='tight')
plt.savefig("result/eshikhon2.svg",dpi=300, bbox_inches='tight')
plt.savefig("result/eshikhon2.jpg",dpi=300, bbox_inches='tight')
plt.show()
#fig.tight_layout()

plt.show()
```
