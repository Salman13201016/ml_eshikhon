```python
data = data.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = data.drop(columns='Sample')
```

```python
data
```

```python
x = data.iloc[:, :-2].values
y = data.iloc[:, 7].values
```

```python
labelencoder = LabelEncoder()
```

```python
x[:, 0] = labelencoder.fit_transform(x[:, 0])
```

```python
x1 = pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=0)
```

```python
len(x_train)
```

```python
from sklearn.model_selection import train_test_split, KFold
```

```python
k = 5
```

```python
kf = KFold(n_splits=k, shuffle=True, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
sc_X = StandardScaler()
```

```python
x_train = sc_X.fit_transform(x_train)
```

```python
x_test = sc_X.transform(x_test)
```

```python
sc_y = StandardScaler()
```

```python
y_train = sc_y.fit_transform(y_train.reshape(-1,1))
```

```python
y_test = sc_y.fit_transform(y_test.reshape(-1,1))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
mse_scores = []
r2_scores = []
for train_index, val_index in kf.split(x_train):
    train_data, val_data = x_train[train_index], x_train[val_index]
    train_target, val_target = y_train[train_index], y_train[val_index]
    
    # Initialize and train a linear regression model
    model = LinearRegression()
    model.fit(train_data, train_target)
    
    # Make predictions on the validation set
    val_predictions = model.predict(val_data)
    
    # Calculate the mean squared error (MSE) for this fold
    mse = mean_squared_error(val_target, val_predictions)
    mse_scores.append(mse)
    r2 = r2_score(val_target, val_predictions)
    r2_scores.append(r2)
```

```python
# Plot the actual vs. predicted values
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(mse_scores) + 1), mse_scores, marker='o', linestyle='-')
plt.plot(range(1, len(r2_scores) + 1), r2_scores, marker='*', linestyle='-')
plt.xlabel('Fold Number')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE & R-2 Scores for Each Fold in k-Fold Cross-Validation')
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
testing_data_model_score = model.score(x_test, y_test)
print("Model Score/Performance on Testing data",testing_data_model_score)

training_data_model_score = model.score(x_train, y_train)
print("Model Score/Performance on Training data",training_data_model_score)
```
